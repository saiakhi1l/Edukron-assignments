```python
# Question 1: Product of first 10 natural numbers
product = 1
for i in range(1, 11):
    product *= i
print(product)
```

```python
# Question 2: Remainder when 156 is divided by 7
print(156 % 7)
```

```python
# Question 3: Square of 25
print(25 ** 2)
```

```python
# Question 4: Cube root of 125
print(125 ** (1/3))
```

```python
# Question 5: Sum of digits in number 12345
num = 12345
digit_sum = sum(int(d) for d in str(num))
print(digit_sum)
```

```python
# Question 6: Check if 97 is a prime number
num = 97
is_prime = True
for i in range(2, int(num**0.5) + 1):
    if num % i == 0:
        is_prime = False
        break
print(is_prime)
```

```python
# Question 7: Factorial of 8
fact = 1
for i in range(1, 9):
    fact *= i
print(fact)
```

```python
# Question 8: Average of numbers: 15, 23, 31, 42, 56
nums = [15, 23, 31, 42, 56]
print(sum(nums) / len(nums))
```

```python
# Question 9: GCD of 48 and 36
def gcd(a, b):
    while b != 0:
        a, b = b, a % b
    return a

print(gcd(48, 36))
```

```python
# Question 10: Sum of first 20 odd numbers
odd_sum = sum(2*i - 1 for i in range(1, 21))
print(odd_sum)
```
