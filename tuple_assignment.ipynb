```python
# Question 5: Count how many times 5 appears in (1, 5, 2, 5, 3, 5, 4, 5, 6)
x=(1, 5, 2, 5, 3, 5, 4, 5, 6)
counter=sum(1 for i in x if i==5)
print(counter)
```

```python
# Question 9: Create a tuple of first 5 even numbers
x=()
for i in range(1,20):
  if i%2==0 and len(x)<=4:
    x=x+(i,)
print(x)
```

```python
# Question 10: Find the average of numbers in tuple (15, 23, 31, 42, 56, 78)
x=(15, 23, 31, 42, 56, 78)
y=sum(x)
z=len(x)
avg=y//z
print(avg)
```
